### Peak Hours of Sales

The busiest hours for pizza sales are between 12 and 13 hours which falls between lunch or brunch meal

```python
df7["hour"] = df7["time"].dt.hour
```

```python
sales_per_hour = df7.groupby("hour")["total_revenue"].sum()
```

```python
sales_per_hour
```

```python
ascendging_sales = sales_per_hour.sort_values(ascending=False)
```

```python
peak_hour = sales_per_hour.idxmax()
```

```python
peak_hour
```

```python
sales_per_hour = df7.groupby("hour")["total_revenue"].count().reset_index()

sales_per_hour
```

```python
plt.plot(sales_per_hour["hour"], sales_per_hour["total_revenue"] , color="g", marker="s", mec="brown", mfc="brown")
plt.title("Total Sales by Hour of the Day", c="brown", fontsize=13)
plt.xlabel("Hour of the Day")
plt.ylabel("Total Sales")
plt.tight_layout()

plt.show()
```

**Note**
- The plt.tight_layout() is used in matplotlib to automatically adjust spacing between elements so they don't overlap or get cut off
- With it, the layout look cleaner, element are properly spaced, and plot is more readable

### Sales Made in Each Month

- There is an increase in sales during july. 
- A downtrend in seen in the month of october, probably due to need for customer engagement or seasonality.

```python
df7["sales_per_month"] = df7["date"].dt.strftime("%B")
```

```python
df7.head()
```

```python
sales_made_each_month = pd.DataFrame(df7.groupby("sales_per_month")["total_revenue"].sum())
```

```python
sales_made_each_month = sales_made_each_month.reset_index()
```

```python
sales_made_each_month
```

```python
sales_made_each_month.columns
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
sales_made_each_month["sales_per_month"] = pd.Categorical(sales_made_each_month["sales_per_month"], categories=month_order, ordered=True)
```

```python
sales_made_each_month = sales_made_each_month.sort_values(by='sales_per_month')
```

```python
sales_made_each_month.head(12)
```

```python
sales_made_each_month
```

```python
plt.plot(sales_made_each_month["sales_per_month"], sales_made_each_month["total_revenue"], marker='*',mec="r", mfc="r")
plt.xticks(rotation=90)
plt.title("Monthly Sales Trend", fontsize= 13, color="brown")
plt.xlabel("Month", c="b")
plt.ylabel("Total Sales", c="b")
plt.tight_layout()

plt.show()
```

- The sales spike during july (July has the highest sales)
- There is a steady decline in sales within august and september

```python
plt.bar(sales_made_each_month["sales_per_month"], sales_made_each_month["total_revenue"], color="skyblue")
plt.xticks(rotation=90)
plt.title("Monthly Sales on Barchart", fontsize= 13, color="b")
plt.xlabel("Month", c="brown")
plt.ylabel("Total Sales", c="brown")

plt.show()
```
